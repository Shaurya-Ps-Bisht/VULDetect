```python
import pandas as pd

def load_data_from_csv(csv_path, num_rows=None):
    
    df = pd.read_csv(csv_path, nrows=num_rows)
    X_train = df['func'].tolist()
    y_train = df['target'].tolist()

    return X_train, y_train
```

```python
csv_path = '/home/shaurya/BTP/1/Outputs/CSV/diversevul.csv'
num_rows = 1000  

X_train, y_train = load_data_from_csv(csv_path, num_rows)

```
